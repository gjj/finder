```python
import pandas as pd
import json
```

```python
input_file = "sgpgrid-api-data-fintech-description"
# df = pd.read_json("{}.json".format(input_file))

with open("{}.json".format(input_file)) as api_results:
    data = json.load(api_results)
    rows = data['companies']['rows']
    df = pd.DataFrame.from_dict(rows)
    
    print(df)
    
    df.to_csv("{}-formatted.csv".format(input_file), index=False)
```
